```python
metrics_to_plot = ["mean", "median", "std", "dark_ratio"]


def fast_read_resize_tif_as_uint8(path: Path, new_size: int = None) -> np.ndarray:
    img = tiff.imread(str(path))  # likely uint16

    # fast downsample by integer factor (no interpolation)
    if new_size is not None:
        factor = img.shape[0] // new_size
        img = img[::factor, ::factor]

    # compute percentiles on the small image, clip, then linearly scale in float32 (avoid float64)
    # do floats in float32 to reduce memory pressure
    p_low, p_high = np.percentile(img, (1, 99))
    img = (img.astype(np.float32) - float(p_low)) * \
        (255.0 / float(p_high - p_low))
    img = np.clip(img, 0, 255)
    return img.astype(np.uint8)


def plot_images(image_paths: list[Path]):
    _, axes = plt.subplots(1, len(image_paths), figsize=(15, 3))

    for ax, path in zip(axes, image_paths):
        ax.imshow(fast_read_resize_tif_as_uint8(
            path, new_size=512), cmap="gray")
        ax.set_title(f"{path.parent.name} {path.name}", fontsize=8)
        ax.axis("off")

    plt.tight_layout()
    plt.show()
    plt.close()


def compute_single_image_metrics(img_arr: np.ndarray):
    mean = img_arr.mean()
    median = np.median(img_arr)
    std = img_arr.std()
    min_val = img_arr.min()
    max_val = img_arr.max()

    p5 = np.percentile(img_arr, 5)
    dark_ratio = (img_arr <= p5).mean()

    return {
        "mean": mean,
        "median": median,
        "std": std,
        "min": min_val,
        "max": max_val,
        "dark_ratio": dark_ratio
    }


def process_path(args):
    path, new_size = args
    img = fast_read_resize_tif_as_uint8(path, new_size)
    return compute_single_image_metrics(img)


# , df_summary: pd.DataFrame):
def output_subsample_metrics(df_metrics: pd.DataFrame, metrics_ranges: Optional[dict] = None):
    # print(df_summary)

    _, axes = plt.subplots(1, 4, figsize=(16, 3))
    for ax, col in zip(axes, df_metrics[metrics_to_plot].columns):
        sns.histplot(df_metrics[col], bins=20, kde=True, ax=ax)
        ax.set_title(col, fontsize=10)

        if metrics_ranges and col in metrics_ranges:
            xlim_min, xlim_max = metrics_ranges[col]
            ax.set_xlim(xlim_min, xlim_max)

            # check if there are outliers beyond robust percentile based axis limits
            min_val = min(df_metrics[col].min(), xlim_min)
            max_val = max(df_metrics[col].max(), xlim_max)

            if min_val < xlim_min or max_val > xlim_max:
                # extend x-axis to show all data
                ax.set_xlim(min_val, max_val)

                # indicate this is an outlier plot
                ax.set_facecolor("#fff0f0")  # light red background
                ax.annotate("Outlier range", xy=(0.5, 0.95), xycoords='axes fraction',
                            ha='center', fontsize=8, color='red')

    plt.tight_layout()
    plt.show()
    plt.close()


def compute_subsample_metrics(image_paths: list[Path], new_size=512):
    args = [(p, new_size) for p in image_paths]

    with ThreadPoolExecutor(max_workers=min(8, (os.cpu_count() or 4))) as exe:
        metrics_list = list(exe.map(process_path, args))

    df_metrics = pd.DataFrame(metrics_list)
    df_summary = df_metrics.describe().T
    output_subsample_metrics(df_metrics)  # , df_summary)

    return df_metrics, df_summary


'''
def compute_global_metrics_ranges(all_metrics: dict[Path, pd.DataFrame]):
    ranges = {}
    for metric in metrics_to_plot:
        all_values = np.concatenate(
            [df[metric].values for df in all_metrics.values()])
        p2, p98 = np.percentile(all_values, [2, 98])
        if metric != "dark_ratio":
            p2 = int(np.floor(p2 / 5) * 5)
            p98 = int(np.ceil(p98 / 5) * 5)
        ranges[metric] = (p2, p98)

    for metric, range in ranges.items():
        print(f"{metric}: {str(range[0])} - {str(range[1])}")
    
    return ranges
'''


def compute_global_metrics_ranges(all_metrics: dict[Path, pd.DataFrame]):
    ranges = {}
    for metric in [m for m in metrics_to_plot if m != "dark_ratio"]:
        all_values = np.concatenate(
            [df[metric].values for df in all_metrics.values()])
        ranges[metric] = (all_values.min(), all_values.max())
    return ranges
```

```python
import re


def plot_img_soi(image_path: Path, xloc: float, yloc: float, tile_x: int, tile_y: int):
    print(
        f"SOI data: xloc={xloc} {type(xloc)}, yloc={yloc} {type(yloc)}, tile_x={tile_x} {type(tile_x)}, tile_y={tile_y} {type(tile_y)}")

    img = fast_read_resize_tif_as_uint8(image_path)
    height_orig, width_orig = img.shape[:2]

    tile_width = 2.0522511476883665e-06

    tile_bl_x = (tile_x - 1) * 0.9 * tile_width
    tile_bl_y = (tile_y - 1) * 0.9 * tile_width

    loc_x_in_tile = xloc - tile_bl_x
    loc_y_in_tile = yloc - tile_bl_y

    scale = width_orig / tile_width

    pixel_x_from_bl = loc_x_in_tile * scale
    pixel_y_from_bl = loc_y_in_tile * scale

    pixel_x = pixel_x_from_bl
    pixel_y = height_orig - pixel_y_from_bl

    pixel_x_backrotated = height_orig - pixel_y
    pixel_y_backrotated = pixel_x

    in_bounds = (0 <= pixel_x < width_orig) and (0 <= pixel_y < height_orig)

    print(f"SOI Pixel Coordinates (rotated): ({pixel_x:.1f}, {pixel_y:.1f})")
    print(f"SOI Pixel Coordinates (back-rotated): ({pixel_x_backrotated:.1f}, {pixel_y_backrotated:.1f})")

    img_np_rotated = np.rot90(img, k=1)
    height_rot, width_rot = img_np_rotated.shape[:2]

    img_norm = (img - img.min()) / (img.max() - img.min() + 1e-8)
    img_rot_norm = (img_np_rotated - img_np_rotated.min()) / \
        (img_np_rotated.max() - img_np_rotated.min() + 1e-8)

    fig, ax = plt.subplots(1, 2, figsize=(12, 12))
    ax[0].imshow(img_rot_norm, cmap='gray', extent=[
                 0, width_rot, height_rot, 0])
    ax[1].imshow(img_norm, cmap='gray', extent=[0, width_rot, height_rot, 0])

    if in_bounds:
        ax[0].plot(pixel_x, pixel_y, 'ro', markersize=25,
                   fillstyle='none', markeredgewidth=2)
        ax[1].plot(pixel_x_backrotated, pixel_y_backrotated, 'ro', markersize=25,
                   fillstyle='none', markeredgewidth=2)

    plt.tight_layout()
    plt.show()
    plt.close()


def extract_tileset_and_coords(s: str):
    pattern = re.compile(
        r"_TileSet_(\d+)_Tile_(\d{3})-(\d{3})-"
    )

    match = pattern.search(s)
    if not match:
        raise ValueError("String does not match expected format")

    tileset = int(match.group(1))
    tile_x = int(match.group(2))
    tile_y = int(match.group(3))

    return tileset, tile_x, tile_y


import pandas as pd


def find_first_matching_row(
    df: pd.DataFrame,
    tileset: int,
    tile_x: int,
    tile_y: int,
):
    tileset_str = f"Tile Set ({tileset})"
    match = df[

        (df["Tileset"] == tileset_str) &
        (df["TileX"] == tile_x) &
        (df["TileY"] == tile_y)
    ]

    if match.empty:
        return None

    return match.iloc[0]


def get_soi_data_for_tif(tif_path: Path, df: pd.DataFrame):
    tileset, tile_x, tile_y = extract_tileset_and_coords(tif_path.stem)
    row = find_first_matching_row(df, tileset, tile_x, tile_y)
    if "LocX" not in row or "LocY" not in row:
        return None, None, None, None
    return float(row["LocX"]), float(row["LocY"]), int(tile_x), int(tile_y)


soi_dirs_annotationfiles = [
    ("Z:\\imcrdata\\@Niklas\\Andrea_Red\\positive",
     "Z:\\imcrdata\\2024_Andrea_NBS1\\2024_Andrea_NBS1_R1\\20240407_Andrea_Red\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Veronica_Red\\positive",
     "Z:\\imcrdata\\2025_Veronica_Lamin\\2024_Veronica_Lamin\\2024_Veronica_Lamin_R2\\20240807_Veronica_Red\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Andrea_Blue\\positive",
     "Z:\\imcrdata\\2024_Andrea_NBS1\\2024_Andrea_NBS1_R2\\20240306_Andrea_Blue\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Andrea_Green\\positive",
     "Z:\\imcrdata\\2024_Andrea_NBS1\\2024_Andrea_NBS1_R2\\20240613_Andrea_Green\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Andrea_Orange\\positive",
     "Z:\\imcrdata\\2024_Andrea_NBS1\\2024_Andrea_NBS1_R1\\20240523_Andrea_Orange\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Andrea_Yellow\\positive",
     "Z:\\imcrdata\\2024_Andrea_NBS1\\2024_Andrea_NBS1_R1\\20240912_Andrea_Yellow\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Dani_Funghi\\positive",
     "Z:\\imcrdata\\2025_Dani_SMC1\\R1\\20250311_Dani_Funghi\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Filip_Appenzell\\positive",
     "Z:\\imcrdata\\2025_Filip_RNF20\\2025_Filip_RNF20_R1\\20250210_Filip_Appenzell\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Filip_Basel\\positive",
     "Z:\\imcrdata\\2025_Filip_RNF20\\2025_Filip_RNF20_R2\\20250108_Filip_Basel\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Filip_Bern\\positive",
     "Z:\\imcrdata\\2025_Filip_RNF20\\2025_Filip_RNF20_R3\\20250728_Filip_Bern\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Filip_Luzern\\positive",
     "Z:\\imcrdata\\2025_Filip_RNF20\\2025_Filip_RNF20_R3\\20250729_Filip_Luzern\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Vero_Larch\\positive",
     "Z:\\imcrdata\\2025_Veronica_Lamin\\2025_Veronica_G9ai\\R2\\20250605_Vero_Larch\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Veronica_Blue\\positive",
     "Z:\\imcrdata\\2025_Veronica_Lamin\\2024_Veronica_Lamin\\2024_Veronica_Lamin_R2\\20240813_Veronica_Blue\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Veronica_Green\\positive",
     "Z:\\imcrdata\\2025_Veronica_Lamin\\2024_Veronica_Lamin\\2024_Veronica_Lamin_R2\\20240820_Veronica_Green\\AnnotationsData.csv"),
    ("Z:\\imcrdata\\@Niklas\\Veronica_Maple\\positive", "Z:\\imcrdata\\2025_Veronica_Lamin\\2025_Veronica_G9ai\\R2\\20250702_Veronica_Maple\\AnnotationsData.csv")]

for dir_str, annotation_file in soi_dirs_annotationfiles:
    dir = Path(dir_str)

    tif_files = random.sample(list(dir.rglob("*.tif")), k=10)

    df_annotations = pd.read_csv(annotation_file)
    for tif_path in tif_files:
        xloc, yloc, tile_x, tile_y = get_soi_data_for_tif(
            tif_path, df_annotations)
        if xloc is not None:
            print(tif_path)
            plot_img_soi(tif_path, xloc, yloc, tile_x, tile_y)
```
